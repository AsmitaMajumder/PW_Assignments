### 1 : What is Dimensionality Reduction? Why is it important in machine

###Question 4:Explain the concept of anomaly detection and its significance.

Answer:
Anomaly detection refers to identifying data points or patterns that deviate significantly from the norm. These unusual observations are called anomalies or outliers.

Significance:

Helps detect fraud, network intrusions, equipment failures, and unusual behavior.

Improves system reliability and security.

###Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:

Statistical Methods:
Use probability distributions or statistical tests to identify points that deviate from expected behavior.

Machine Learning-Based Methods:
Use algorithms like Isolation Forest, One-Class SVM, or Autoencoders to learn normal patterns and detect abnormalities.

Distance or Density-Based Techniques:
Identify anomalies by measuring distances or density differences (e.g., k-NN, DBSCAN).

###Question 6: What is time series analysis? Mention two key components of time series
data.

Answer:
Time series analysis involves studying data points collected or recorded at specific time intervals to identify patterns, trends, and future behavior.

Two key components:

Trend: Long-term upward or downward movement in the data.

Seasonality: Repeating patterns occurring at regular intervals

###Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:

Seasonality:
Patterns that repeat at fixed, known intervals (e.g., daily, monthly, yearly). They are predictable.

Cyclic Behavior:
Patterns that occur over irregular, long-term cycles influenced by external conditions (e.g., economic cycles). They are not periodic or strictly predictable.

###Question 8: Write Python code to perform K-means clustering on a sample dataset.


```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plot output
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("K-Means Clustering Output")
plt.show()
```

###Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:
Inheritance is an object-oriented programming concept where one class (child/subclass) acquires the properties and methods of another class (parent/superclass). It promotes code reuse and hierarchy.

```python
class Animal:
    def speak(self):
        return "This animal makes a sound."

class Dog(Animal):
    def speak(self):
        return "The dog barks."

# Usage
d = Dog()
print(d.speak())   # Output: The dog barks.
```
