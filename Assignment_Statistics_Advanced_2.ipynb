### Q1.What is hypothesis testing?

Ans:
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses (null and alternative) and using sample evidence to determine which hypothesis is more likely to be true. The test provides a systematic way to evaluate claims using probability theory.

###Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?



Null Hypothesis (H₀): A statement that assumes no effect, no difference, or no relationship exists. It represents the default or status quo assumption.

Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis and suggests there is an effect, difference, or relationship.
Difference: The null hypothesis assumes no change, while the alternative hypothesis proposes a significant change or effect.

###Q3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 (5%) or 0.01 (1%).

If the p-value ≤ α → Reject H₀ (evidence supports H₁).

If the p-value > α → Fail to reject H₀ (insufficient evidence against H₀).
It sets the threshold for decision-making and controls the risk of Type I error

###Q4.What are Type I and Type II errors? Give examples of each.

Answer:

Type I Error (False Positive): Rejecting H₀ when it is true.
Example: A medical test indicates a person has a disease when they are healthy.

Type II Error (False Negative): Failing to reject H₀ when it is false.
Example: A medical test fails to detect a disease when the person actually has it.

###Q5.What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

Z-test: Used when the population variance is known and the sample size is large (n > 30). Based on the standard normal distribution.

T-test: Used when the population variance is unknown and/or the sample size is small (n ≤ 30). Based on the Student’s t-distribution.
When to use:

Z-test: Large sample, known variance.

T-test: Small sample, unknown variance.

###Q6.: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000  # number of experiments

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```


##Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu = 50
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (mean - mu) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

###Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
data = np.random.normal(loc=50, scale=5, size=1000)

# Confidence interval (95%)
mean = np.mean(data)
std_err = stats.sem(data)
conf_interval = stats.norm.interval(0.95, loc=mean, scale=std_err)

# Plot
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution with 95% CI")
plt.axvline(conf_interval[0], color='red', linestyle='--', label=f"95% CI lower: {conf_interval[0]:.2f}")
plt.axvline(conf_interval[1], color='red', linestyle='--', label=f"95% CI upper: {conf_interval[1]:.2f}")
plt.legend()
plt.show()

print("95% Confidence Interval:", conf_interval)
```

###Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Example dataset
data = np.random.normal(50, 10, 100)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram
plt.hist(z_scores, bins=20, edgecolor='black')
plt.title("Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("First 10 Z-scores:", z_scores[:10])
```
